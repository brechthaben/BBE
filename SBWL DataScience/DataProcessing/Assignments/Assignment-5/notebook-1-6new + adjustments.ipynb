```python
import pandas as pd
import geopandas as gpd
import json
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pathlib

print("Starting streamlined script...")

# ==============================================================================
# Configuration & Setup
# ==============================================================================
CWD = pathlib.Path(os.getcwd())
DATA_DIR = CWD / "data"
CRIME_JSON_PATH = DATA_DIR / "data_notebook-notebook-1_dataset2.json"
CENTROIDS_CSV_PATH = DATA_DIR / "CA_tract_centroids_2020.csv.txt"
INCOME_CSV_PATH = DATA_DIR / "data_notebook-notebook-1_dataset1.csv"
PROJECTED_CRS = "EPSG:3310"  # NAD83 / California Albers
NUM_TOP_TRACTS = 5
NUM_TOP_CRIMES_TO_PLOT = 5
YEARS_TO_PLOT = range(2013, 2020)

if not DATA_DIR.is_dir():
    raise FileNotFoundError(f"Data directory '{DATA_DIR}' not found.")

# ==============================================================================
# PART 1: Load Income & Compute Score Change
# ==============================================================================
print("\n--- Processing Financial & Geographic Data ---")

# Load Centroids
centroids_df = pd.read_csv(CENTROIDS_CSV_PATH)
centroids_df['geoid'] = (
    centroids_df['STATEFP'].astype(str).str.zfill(2) +
    centroids_df['COUNTYFP'].astype(str).str.zfill(3) +
    centroids_df['TRACTCE'].astype(str).str.zfill(6)
)
centroids_lookup_df = centroids_df[['geoid', 'LATITUDE', 'LONGITUDE']].copy()

# Load Income Data
df_income = pd.read_csv(INCOME_CSV_PATH)
# Compute exclusive percentages
df_income['ex_eli_pct']   = df_income['eli_pct']
df_income['ex_vli_pct']   = df_income['vli_pct'] - df_income['eli_pct']
df_income['ex_li_pct']    = df_income['li_pct'] - df_income['vli_pct']
df_income['ex_mi_pct']    = df_income['mi_pct'] - df_income['li_pct']
df_income['ex_abmi_pct']  = df_income['abmi_pct']
# Weighted financial score
weights = {'ex_eli_pct': -1.0, 'ex_vli_pct': -0.5, 'ex_li_pct': 0.0, 'ex_mi_pct': 0.5, 'ex_abmi_pct': 1.0}
df_income['financial_score'] = sum(df_income[col] * w for col, w in weights.items()) / 100
# Compute first vs. last year score and the change
df_income['year'] = pd.to_numeric(df_income['year'], errors='coerce').astype(int)
df_sorted = df_income.sort_values(['geoid', 'year'])
first_scores = df_sorted.loc[df_sorted.groupby('geoid')['year'].idxmin()]
last_scores  = df_sorted.loc[df_sorted.groupby('geoid')['year'].idxmax()]
score_change = pd.merge(
    first_scores[['geoid','financial_score','year']].rename(columns={'financial_score':'first_score','year':'first_year'}),
    last_scores [['geoid','financial_score','year']].rename(columns={'financial_score':'last_score','year':'last_year'}),
    on='geoid'
)
score_change = score_change[score_change['first_year'] != score_change['last_year']].copy()
score_change['change'] = score_change['last_score'] - score_change['first_score']
# Get top tracts
top_changers = score_change.reindex(score_change['change'].abs().sort_values(ascending=False).index).head(NUM_TOP_TRACTS)
top_changers['geoid_clean'] = top_changers['geoid'].astype(str).str.extract(r'(\d{11})').fillna(top_changers['geoid'])
top_changers_lookup = top_changers.set_index('geoid_clean')[['change','first_year','last_year']]
top_geoids = top_changers_lookup.index.tolist()

print(f"Identified top {len(top_geoids)} tracts by financial change: {top_geoids}")

# ==============================================================================
# PART 2: Load Crime Data & Spatial Join
# ==============================================================================
print("\n--- Processing Crime Data & Performing Spatial Join ---")

# Load raw crime JSON
with open(CRIME_JSON_PATH, 'r') as f:
    try:
        crime_data_list = json.load(f)
    except json.JSONDecodeError:
        crime_data_list = [json.loads(line) for line in f if line.strip()]

crime_df = pd.DataFrame(crime_data_list)
# Clean coordinate columns
crime_df.rename(columns={c: c.lower() for c in crime_df.columns}, inplace=True)
crime_df['lat'] = pd.to_numeric(crime_df['lat'], errors='coerce')
crime_df['lon'] = pd.to_numeric(crime_df['lon'], errors='coerce')
crime_df.dropna(subset=['lat','lon'], inplace=True)

# Build GeoDataFrames
crime_gdf      = gpd.GeoDataFrame(crime_df, geometry=gpd.points_from_xy(crime_df.lon, crime_df.lat), crs="EPSG:4326")
centroids_gdf  = gpd.GeoDataFrame(centroids_lookup_df, geometry=gpd.points_from_xy(centroids_lookup_df.LONGITUDE, centroids_lookup_df.LATITUDE), crs="EPSG:4326")

crime_proj     = crime_gdf.to_crs(PROJECTED_CRS)
centroids_proj = centroids_gdf.to_crs(PROJECTED_CRS)
centroids_proj = centroids_proj.reset_index(drop=True)

# Spatial join to tag each crime with the nearest tract geoid
crimes_joined = gpd.sjoin_nearest(crime_proj, centroids_proj[['geometry']], how='left')
crimes_with_geo = crimes_joined.merge(
    centroids_proj[['geoid']],
    left_on='index_right',
    right_index=True,
    how='left'
)

# Extract year and filter to top_geoids
crimes_with_geo['date_occ'] = pd.to_datetime(crimes_with_geo['date_occ'], errors='coerce')
crimes_with_geo.dropna(subset=['date_occ'], inplace=True)
crimes_with_geo['year'] = crimes_with_geo['date_occ'].dt.year
crimes_with_geo['crm_cd_desc'] = crimes_with_geo['crm_cd_desc'].fillna('UNKNOWN').str.strip()
crimes_top = crimes_with_geo[crimes_with_geo['geoid'].isin(top_geoids)].copy()

# Merge in the financial change for each crime record (for later use)
crimes_top = crimes_top.merge(
    top_changers_lookup[['change']],
    left_on='geoid',
    right_index=True,
    how='left'
)

print(f"Filtered down to {len(crimes_top)} crime records in the selected tracts.")

# ==============================================================================
# PART 3: Precompute Pivot of Crime Counts per Tract & Year
# ==============================================================================
all_counts = (
    crimes_top
    .groupby(['geoid','year','crm_cd_desc'])
    .size()
    .reset_index(name='count')
    .pivot_table(index=['geoid','year'], columns='crm_cd_desc', values='count', fill_value=0)
)

# ==============================================================================
# PART 4: Plot Combined Visualization for Each Tract
# ==============================================================================
print(f"\n--- Generating Combined Plots for Top {len(top_geoids)} Tracts ---")

for geoid in top_geoids:
    if geoid not in all_counts.index.get_level_values('geoid'):
        print(f"Skipping GEOID {geoid}: no crime data")
        continue

    # Extract yearly crime counts for this tract
    crime_pivot = all_counts.loc[geoid].reindex(YEARS_TO_PLOT, fill_value=0)

    # Fetch financial score change for title
    change_info = top_changers_lookup.loc[geoid]
    change_text = f"Financial Score Change ({int(change_info['first_year'])}-{int(change_info['last_year'])}): {change_info['change']:+.2f}"

    # Plot
    plt.figure(figsize=(12,7))
    colors = sns.color_palette('tab10', n_colors=len(crime_pivot.columns))
    for i, crime_type in enumerate(crime_pivot.columns):
        plt.plot(crime_pivot.index, crime_pivot[crime_type],
                 marker='o', linestyle='-', label=crime_type, color=colors[i])

    plt.title(f"Crime Trends & Financial Change – GEOID {geoid}\n{change_text}", fontsize=14)
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Reported Incidents", fontsize=12)
    plt.xticks(list(YEARS_TO_PLOT))
    plt.grid(True, axis='y', linestyle='--', alpha=0.6)
    plt.legend(title='Crime Type', loc='center left', bbox_to_anchor=(1.02, 0.5))
    plt.tight_layout(rect=[0,0,0.85,1])
    sns.despine()
    plt.show()

print("\n--- Script finished ---")
```
